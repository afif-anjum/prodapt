## Our final data contains only relevant column with 0 missing values

### Step 1 -Evaluate Phy Rates at Different RSSI levels

#### 1. Look at the RSSI levels (rssi_percs_25) and look at the spread of Tx and Rx weighted Phy Rate

```python
#scatter plot to check the spread of Tx and Rx weighted Phy rate across rssi_percs_25
plt.figure(figsize=(20,15))
plt.subplot(2,1,1)
sns.scatterplot(x=final_df['rssi_percs_25'],y=final_df['txWeightedPhyRate'])

plt.subplot(2,1,2)
sns.scatterplot(x=final_df['rssi_percs_25'],y=final_df['rxWeightedPhyRate'])

plt.show()
```

#### 2. Produce a histogram (graph and CSV) for the average Tx and Rx Weighed Phy Rate between -85 and -65 dB in 1 dB steps 

```python
#filtering rows where rssi_percs_25 is between -85 and -65
temp_df = final_df[(final_df['rssi_percs_25']>=-85) & (final_df['rssi_percs_25']<=-65)]

#calculating average tx and rx weighted phy rate across rssi_percs_25
temp_df = temp_df.groupby("rssi_percs_25").agg({'txWeightedPhyRate':'mean','rxWeightedPhyRate':'mean'}).reset_index()
temp_df = temp_df.rename(columns = {"txWeightedPhyRate":"avg(txWeightedPhyRate)","rxWeightedPhyRate":"avg(rxWeightedPhyRate)"})
temp_df.to_csv('average_tx_and_rx_phy_rates_across_rssi.csv',index=False)
```

```python
#bar plot to show the histogram
plt.figure(figsize=(20,15))

plt.subplot(2,1,1)
sns.barplot(x=temp_df['rssi_percs_25'],y=temp_df['avg(txWeightedPhyRate)'])
plt.title("average of txWeightedPhyRate vs rssi_percs_25")

plt.subplot(2,1,2)
sns.barplot(x=temp_df['rssi_percs_25'],y=temp_df['avg(rxWeightedPhyRate)'])
plt.title("average of rxWeightedPhyRate vs rssi_percs_25")

plt.show()
```

## Step 2: Evaluate RSSI threshold

### 1. % of devices connected to 2.4 GHz and 5 GHz

```python
temp = final_df.groupby("freq_band")['client_mac'].nunique().sort_values(ascending = False).reset_index()
temp.rename(columns = {'client_mac': 'total_devices'},inplace=True)
temp['device_%age'] = round(temp['total_devices']/temp['total_devices'].sum()*100,0)
temp
```

*Finding* : 51% of total devices are connected to 2.4G where as 49% of devices are connected to 5G

### 2. Evaluating new freq basis business rule

```python
def get_new_freq_band(row):
    if ((row['freq_band']=='2.4G') and (row['rssi_percs_25'] >= -60)):
        return '5G'
    elif ((row['freq_band']=='5G') and (row['rssi_percs_25'] <= -75)):
        return '2.4G'
    elif ((row['freq_band']=='5G') and (row['rssi_percs_25'] > -75)):
        return '5G'
    elif ((row['freq_band']=='2.4G') and (row['rssi_percs_25'] < -60)):
        return '2.4G'
final_df['new_freq_band'] = final_df.apply(get_new_freq_band,axis=1)

final_df.head()
```

```python
temp = final_df.groupby("new_freq_band")['client_mac'].nunique().sort_values(ascending = False).reset_index()
temp.rename(columns = {'client_mac': 'total_devices'},inplace=True)
temp['device_%age'] = round(temp['total_devices']/temp['total_devices'].sum()*100)
temp
```

*Findings* : Post making the transition to new_freq_band, 41% of total devices are connected to 2.4G whereas for 5G band %age connected device increased to 59%

Yes, Post transition there will be more devices connected to 5G
